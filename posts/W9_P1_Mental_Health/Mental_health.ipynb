# AI for Mental Health
I am curious about how LLM may help with mental health or emotional situations, so I tested how gpt and gemini perform when a user sharing anxiety. I also compare between prompts. 

When I put 
> “I feel really stressed. I'm worrying about a lot of things...”




## Chatgpt response :

> I’m really sorry you’re feeling this way. When stress piles up like that, it can feel overwhelming, like everything is pressing at once. You don’t have to sort it all out right now—and you don’t have to go through it alone.

> If you’re up for it, we can slow things down together. Sometimes it helps to gently unpack the stress instead of holding it all at once. A few options—tell me what feels easiest:
Name it: What are the main things you’re worrying about right now? Even a short list is okay.
Zoom in: Is there one worry that feels loudest or most urgent at this moment?
Ground for a minute: We can do a quick 60-second reset (breathing, grounding, or just pausing) before talking more.

> If at any point your stress feels unmanageable or you’re feeling unsafe, it’s important to reach out to someone who can support you right away—like a trusted person in your life or a professional. If you’re in the U.S., you can call or text 988 for the Suicide & Crisis Lifeline, available 24/7.
I’m here with you. What’s been weighing on you the most?

## Gemini Response:

I hear you, and I want to validate that it is completely understandable to feel overwhelmed when worries start to pile up. When everything feels like a priority, it can be paralyzing.
You don’t have to "fix" everything right this second. Here is a gentle way to decompress and regain a little bit of control.

1. The Physical Reset
Before tackling the thoughts, try to lower your body's physical stress response.
The 4-7-8 Breath: Inhale through your nose for 4 seconds. Hold that breath for 7 seconds. Exhale slowly through your mouth (like you’re blowing out a candle) for 8 seconds.
Unclench: Check your jaw, drop your shoulders away from your ears, and unstick your tongue from the roof of your mouth.
2. Get it Out of Your Head
One of the biggest sources of stress is trying to hold onto every worry mentally so you don't forget them.
The "Brain Dump": Take a piece of paper (or use the chat here) and list every single thing you are worried about. Big things, small things, silly things.
Why this works: Once it is written down, your brain feels safe letting go of the active effort to "remember" the worry.
3. Categorize the Chaos
Once you have your list, look at the items and mentally sort them into two buckets:
Bucket A (Out of Control): Things involving other people's reactions, the future, or the past. Acknowledge them, but recognize you cannot act on them right now.
Bucket B (In Control): Things you can actually influence today.
4. The "One Small Thing" Rule
Pick one very small, easy thing from Bucket B. It could be as simple as "reply to one text" or "drink a glass of water." Doing one small thing breaks the paralysis loop.

How can I help you right now?
Would you like to do a "brain dump" right here in the chat? You can list everything out, and I can help you organize it, or I can just listen if you need to vent.


## To Conclude:

From these two responses, it is clear that both ChatGPT and Gemini are designed to respond to emotional distress with empathy, but they take noticeably different approaches in how that empathy is expressed and structured.

ChatGPT’s response feels conversational and open-ended. It starts by validating the emotion and normalizing the experience of stress, then gently invites the user into reflection by offering choices. Phrases like “If you’re up for it” and “What feels easiest” give the user a sense of agency, rather than pushing them toward a specific action. The response also emphasizes that the user does not need to solve everything immediately, which can be comforting in moments of anxiety. Importantly, ChatGPT leaves space for the user to decide how much they want to share, and ends with an open question that encourages continued dialogue. This makes the interaction feel more like a supportive listener than a problem-solver.

Gemini’s response, on the other hand, is much more structured and instructional. After validating the user’s feelings, it quickly shifts into a step-by-step coping framework, including breathing techniques, body awareness, writing exercises, and categorization strategies. The language is still caring, but the response feels closer to a guided self-help exercise. By breaking stress down into physical, cognitive, and behavioral steps, Gemini appears to prioritize helping the user “do something” immediately. This may be especially helpful for users who want concrete actions when they feel overwhelmed, but it could also feel a bit heavy or directive for someone who is simply seeking emotional acknowledgment.

Comparing the two, I noticed that prompt design plays a big role in shaping emotional support responses. With the same user input, ChatGPT leaned toward emotional validation and conversation, while Gemini leaned toward structured intervention. Neither approach is inherently better; they serve different emotional needs. In moments of anxiety, some people may want space to talk and feel heard, while others may want practical steps to regain a sense of control. This experiment made me realize that LLMs are not just responding to what we say, but also implicitly choosing a role such as a listener, a coach, or a guide, based on their design and safety priorities.

Overall, this test made me more aware of how LLMs might support mental health in limited, non-clinical ways. They are not therapists, but they can model empathy, suggest grounding strategies, and reduce the feeling of being alone. At the same time, their responses reflect design choices about agency, responsibility, and risk management. Understanding these differences is important, especially as people increasingly turn to AI during emotionally vulnerable moments.
