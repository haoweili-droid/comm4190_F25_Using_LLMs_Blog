# Public Speaking Coach Agent
I have had an idea to create a public speaking coach agent that helps college students and working professionals to proofread projects, practice presentations or pitches and provide personalized comments by taking roles of stakeholders. 

Background: As LLMs are getting increasingly powerful, everyone can create products, yet few have the entrepreneurship skills to sell their products to investors and customers. People need perspectives on their work, but they cannot always find the most helpful resources. This presentation coach App can proof read any project including papers, product document, slide decks, posters, personal statement etc. Users just need to upload or take a photo of the work, and LLMs will play the role of a group of readers to provide evaluations from a particular perspective. The user can interact with the LLM via text, audio chat or even gesture, just like how people in the real world will communicate. LLM will hold role play sessions with the user to help them improve their speech. For example, if a marketing manager wants to proofread her campaign plan to prepare for her presentation with clients tomorrow, the LLM can play the role of specific clients and give targeted feedback that will help the marketing manager improve for the specific task. With this App, LLM can take on different personas to provide reviews. This will look similar to using LLM to do a product AB testing, but with richer details, more comprehensive suggestions on improving project and presentations, and more actionable results.

To test this project idea, I tried uploading a presentation slide deck to gpt 5.2 and gemini 3 to ask for feedback. This slide is created for a product need finding class, and we have already presented to clients and received feedback, allowing me to compare LLMs answers with real world scenarios. 


Here is the prompt I wrote:

> You are marketing and R&D managers at Campbell’s. You are working on your V8 product series and hope to attract more young consumers.

> You will see a presentation deck on a needfinding research a group conducted to investigate vegetable goodness for younger generation (age 18-34)

> Task:
1- evaluate this presentation deck and point out strength and weakness.
2- give 2-3 follow up questions based on the slide
3- offer suggestions for improvement on the slides and presentation

>Output: a list of concise information

## Gemini Response

## Gemini Feedback: Campbell’s Marketing & R&D Perspective

Hello! Speaking as your counterparts in **Marketing and R&D at Campbell’s**, we’ve reviewed your **VitalPress** proposal. We are impressed by the depth of your user empathy and the boldness of the concept. It addresses a critical gap for us: making **V8 relevant to a younger demographic** that values freshness and transparency.

Below is our evaluation of the deck, followed by questions and suggestions to help you refine this for your final presentation.

---

## 1. Evaluation of the Deck

### Strengths (What Works Well)

**Deep Consumer Insight (Needfinding)**  
You moved beyond surface-level observations. Identifying that *“Healthy is an emotion, not a nutrient label”* is a powerful insight. Recognizing that *“convenience is their new self-care”* effectively frames the conflict for busy students.

**Strong Methodology**  
The use of the **AEIOU framework** and **projective exercises** (e.g., sketching ideal products) generated rich qualitative data. Screening for users who *rarely cook but value vegetable intake* ensured you targeted the right “non-consumer.”

**Visual Storytelling**  
The storyboard clearly dramatizes the pain point (stress, hunger, poor choices) and the solution (a fresh reset). The transition from low-fidelity sketches to high-fidelity renders demonstrates a clear and thoughtful design process.

**On-Trend Benefit Framing**  
Shifting from nutrient-based claims (e.g., Vitamin C) to functional emotional states such as **Focus**, **Calm**, and **Energy** aligns well with current Gen Z wellness trends.

---

### Weaknesses (Areas for Growth)

**Operational Feasibility (R&D Perspective)**  
The concept relies on daily restocking of fresh vegetables and a cold-press system inside the machine. From an R&D standpoint, fresh produce spoilage and sanitation of an automated juicer (pulp disposal, pipe cleaning) are major technical hurdles that are not addressed.

**Business Model Clarity (Marketing Perspective)**  
While desirability is clear, viability is underdeveloped. Fresh produce combined with cold-press technology suggests a high price point. Would a student realistically pay \$8–\$10 for a vending machine juice?

**Slide Density**  
Some slides, particularly **Secondary Research** and **Visual Framework**, are text-heavy. In a live presentation, this risks pulling attention away from the presenter.

---

## 2. Follow-Up Questions

As managers, we would ask the following during the Q&A:

**Trust & Transparency**  
You note that students rely on visual cues such as freshness and color to trust a product. If vegetables are hidden until pressed, how do users verify freshness before purchase? Is a transparent window enough, or do they need visibility into the raw inventory?

**V8 Brand Connection**  
Your research shows young consumers prefer clean-label and modern aesthetics. Does the V8 logo build trust, or does it carry legacy baggage (e.g., salty canned juice)? How does VitalPress bridge this perception gap?

**The “Reset” Moment**  
You identify the reset moment as the core opportunity. Is VitalPress intended as a **meal replacement** (satiety) or a **refreshment** (hydration)? This distinction affects calorie messaging and portion sizing.

---

## 3. Suggestions for Improvement

### Refining the Concept (The Product)

**Address the “Messy” Reality**  
To strengthen R&D credibility, consider adjusting the mechanism. Instead of pressing raw vegetables on site, could VitalPress mix **high-potency, flash-frozen vegetable purees**? This reduces spoilage risk while preserving freshness and nutrient density (e.g., via HPP).

**Enhance the App’s Role**  
You mention an app for customization. Emphasize habit formation. Could users subscribe to a 3 PM “Focus” juice so it’s ready after class? This reinforces the ritual aspect uncovered in your research.

---

### Refining the Slides (The Pitch)

**Show, Don’t Tell the Research**  
On the Visual Framework slide, replace some explanatory text with direct interview quotes or observation photos to humanize the data.

**Move the Storyboard Earlier**  
Consider placing the storyboard before the final concept reveal. Let the audience feel the student’s stress and decision fatigue before introducing VitalPress as the solution.

**Simplify the “Why”**  
On Slide 3, multiple statistics dilute impact. Highlight one compelling data point (for example, the TikTok trust statistic) to create a stronger hook.

## Reflection:
In this task, Chatgpt provided answers closer to the actual scenario than Gemini. While Gemini focused on some nuanced details such as logo and branding, Chatgpt posted questions on findings about customer needs, which was exactly the questions we got in real presentation. There were 12 groups at the presentation in that class, but none of them were asked about logos or visual cues. Here, Chatgpt is better at understanding the context, the role of a marketing or L&D manager, and the main requirement of a client presentation. 

Using AI to role-play as Marketing and R&D managers at Campbell’s changed how I evaluated my own work. By forcing the AI to stay in character, I had to step outside the student or designer mindset and think from the perspective of people responsible for brand risk, feasibility, and long-term strategy. This surfaced gaps I had not fully addressed, especially around operational complexity, sanitation, pricing, and scalability. These concerns did not naturally emerge from user interviews, but they are essential in real-world product development. Seeing my concept questioned through this lens helped me understand where strong insights still needed structural support to become viable.

At the same time, the AI persona functioned as a low-stakes rehearsal space for cross-functional critique. Because the feedback came from a simulated role rather than a real stakeholder, I was more open to iteration and less defensive about the idea. This made it easier to refine my argument and anticipate tough questions before a final presentation. Overall, this process showed me that AI is most valuable not as a content generator, but as a perspective-taking tool that helps bridge the gap between research-driven design and business reality.
