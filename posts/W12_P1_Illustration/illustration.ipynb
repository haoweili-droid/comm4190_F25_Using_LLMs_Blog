# Image generation - illustration for children
I am a designer for a children magazine at Penn. With recent models development in image generation, I wanted to try it out for an illustration for an online article. 

## I first asked Chatgpt (gpt 5.2) to generate an image prompt:

**Prompt**
Create a single wide illustration for K8 audience inspired by the article “Seeking a Fresh Start: New Year’s Across the Globe”, visually showing how different cultures around the world celebrate the New Year in unique ways. The illustration should feel unified and global, cartoon style with multiple celebrations happening within one connected scene.

What to Illustrate

Nowruz (Persian New Year): Families and friends jumping over small bonfires and gathering around a festive table with symbolic items like apples, flowers, and coins. Include signs of early spring such as greenery or blooming flowers.

Lunar New Year: A lively celebration with glowing lanterns, colorful lion dances, and families spending time together. Children may be holding red envelopes, but they should be blank with no markings.

Yennayer (Amazigh New Year): A warm winter gathering with people wearing traditional clothing and jewelry, sharing food such as couscous or communal dishes, and enjoying music and togetherness.

The three celebrations should blend naturally into one scene rather than appearing as separate panels.

Style

Any illustrative style or medium is welcome.

Please do NOT use a clean vector or Illustrator-style look.

The mood should feel festive, warm, and respectful of each culture.

Character Rules (Important)

Each person should ONLY have two arms.

Each hand should have exactly five fingers.

No extra limbs or distorted anatomy.

Requirements
╰┈➤ Avoid putting words in your illustration—the articles already have enough!
╰┈➤ Dimensions are 1920 × 1080 pixels.
╰┈➤ Save as PNG or JPEG.

The final illustration should clearly communicate the idea of new beginnings and celebration across cultures, based directly on the events and traditions described in the article.

> I made some modifications to the prompt. For example, I emphasized that each person should only have two arms and each hand should have exactly five fingers because that is the parts where AI will most likely go crazy.

Then I sent the prompt to gpt 5.2, gemini/nanobanana, and Dalle-3. Below are the results:

Personally, I like Nanobanana illustration the best because it captures the feeling I hope to reach and is logical. The one generated had a random Santa like figure playing a musical, and the one Dalle 3 generated just didn’t fit kids audiences. The size was also not as required in the prompt.

However, I do think that all the characters in the three images were typical human figures, so restricting character rule in the prompt was helpful for reducing wired cases in images. Still, I feel that making small modifications and iterations on images are hard using AI. Everytime I just wanted to change a small part on the image, more places need to be fixed in the new version. Therefore, I think updating the prompt and having new conversations every time is helpful for making edits on illustration. 

In general, I think AI works really well at capturing the sense of article and creating an illustration. This will at least be helpful for teachers who do not have extensive design skills to bring more engaging content to their students. However, I wonder how professional designers or illustrators may view the quality of the illustrations and their usage in magazines.
