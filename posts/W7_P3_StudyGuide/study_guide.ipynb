# Am I Still the Author? Ownership in AI-Assisted Work

Defining the Question: What Does Authorship Mean When AI Is Involved?

Authorship has traditionally been tied to originality, intention, and responsibility. If I write an essay, design a lesson, or develop an idea, I am the author because the choices, framing, and final decisions come from me. However, working with large language models complicates this definition. When an AI helps generate language, structure, or even ideas, it raises a simple but uncomfortable question: am I still the author of this work?

This question is not just about credit or plagiarism. It is about voice, agency, and accountability. If an AI helps shape my writing, whose thinking is actually being expressed? And if something is unclear, misleading, or wrong, who is responsible for it? Rather than treating these questions abstractly, I wanted to explore them through direct interaction with LLMs.

## Testing the Question: Writing With and Without AI
To explore authorship, I experimented with using an LLM at different stages of writing. In some cases, I asked the model to generate a full draft based on a rough idea. In others, I wrote a draft myself and used the AI only for revision, clarification, or feedback. I also tested prompts that explicitly asked the model to “write in my voice” versus prompts that were more neutral and open-ended.

As I compared these outputs, I paid attention to how much control I felt over the content. Did the writing feel like something I would say? Did I recognize my own thinking in it? I also looked at how much effort was required to revise the AI-generated text into something I felt comfortable claiming as my own.

Finally, I reflected on my emotional response. In some cases, I felt relief and efficiency. In others, I felt strangely detached from the output, even when it was technically well written.

## Observations

One clear pattern emerged. When the AI generated large portions of content, my sense of authorship weakened. Even if the ideas aligned with my beliefs, the language often felt generic or overly polished. I could edit it, but the core structure did not originate from my own reasoning process. In those moments, I felt more like an editor than an author.
In contrast, when I used the LLM as a thinking partner rather than a generator, my sense of ownership remained strong. Asking the AI to challenge my ideas, suggest counterarguments, or help reorganize my writing felt similar to working with feedback from a peer. The final product still felt like mine because the intellectual direction came from me.
I also noticed that responsibility shifted depending on how I used the model. When I relied heavily on AI-generated explanations, I was more likely to accept claims without fully verifying them. This made me realize that even if AI contributes content, responsibility cannot be outsourced. If my name is on the work, the accountability is still mine.

Reflecting on the Process: Redefining Authorship, Not Losing It
This exploration helped me rethink authorship in AI-assisted work. Instead of asking whether I am still the author, a better question might be: what role am I playing? Authorship, in this context, is less about producing every sentence and more about setting intentions, making judgments, and taking responsibility for the final outcome.
I have come to see AI as something that can either distance me from my work or deepen my engagement with it, depending on how I use it. When I let the model lead, my voice fades. When I use it to support reflection, revision, or exploration, my voice becomes clearer. In this sense, authorship is not lost, but it does require active effort to maintain.
Ultimately, AI has forced me to be more explicit about my own thinking. I can no longer assume that writing automatically reflects understanding or ownership. Instead, I have to ask myself whether I can explain, defend, and stand behind what is written. If I can, then I am still the author, regardless of the tools I used along the way.
Rather than replacing authorship, AI has made it visible. It has turned writing into a series of choices about control, voice, and responsibility. And those choices, more than the presence or absence of AI, are what define authorship for me now.

### Key Concepts Explained

#### 1. The Three Types of Cognitive Load

Think of your brain like a computer's RAM. It has limited space.

- **Intrinsic Load (The "What")**  
  The inherent difficulty of the material itself.  
  *Analogy:* Lifting a 50lb weight. It is heavy because it is heavy.

- **Extraneous Load (The "How")**  
  Unnecessary effort caused by poor design or instruction.  
  *Analogy:* Lifting the same 50lb weight while blindfolded and wearing slippery gloves.

- **Germane Load (The "Learning")**  
  Productive effort used to build schemas and understanding.

## Part IV: Reflection
Overall, the AI-generated study guide generally supports the two ideas that motivated this project, though in a fairly expected and incremental way rather than a transformative one. The guide successfully shifts attention from quiz scores to underlying conceptual understanding by identifying specific misconceptions and organizing review content around them. This suggests that using completed quizzes as input can help students return to the original learning goals of the course, especially when they want to relearn material after an assignment is over. In this sense, the tool does help students get back on track without requiring them to reread everything or start from scratch.

The output also aligns with the idea of using AI to support more personalized and adaptive learning. Instead of reviewing all concepts equally, the study guide focuses on cognitive load distinctions, which were clearly problematic based on the quiz responses. The explanations, analogies, and short practice activities are tailored to those gaps, while concepts the student already understands are largely left alone. This shows how AI can use evidence from student work to prioritize learning needs and generate a targeted study plan, something that is difficult to scale through traditional instruction alone.

From a broader perspective, this result reflects a more balanced role for AI in learning. Rather than acting as a shortcut to answers, the AI functions as a post-assessment support tool that encourages reflection and re-engagement with course concepts. At the same time, the quality of the output depends heavily on how the prompt constrains the AI’s role. Without those constraints, the AI could easily default to giving direct answers. This suggests that AI’s impact on learning is less about the technology itself and more about how intentionally it is framed within the learning process.
