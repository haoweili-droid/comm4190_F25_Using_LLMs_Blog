#  Creating a study guide based on quizzes? Turn shortcut to meaningful learning  
There are many tools on the market that help teachers create assignments and assessments based on curriculum content. However, I am thinking the other way around. Can we create a tool that helps students to re-study a class based on their homework answers? 

The motivation behind this idea include two parts:

1) With LLMs, people sometimes rely on AI to get an answer instead of actually learning the content. That is understandable because AI is just so tempting, and when you are stressed out with all the schoolwork, it is common for someone to offload mental work to AI. The way to learn content (or not) totally depends on one’s own goal. However, what I am concerned about is: what if someone looks back from the assignment and wants to learn again? Is there a way to help them quickly on track with their original learning path?

2) Learning can be personalized and adaptive with AI. If students can upload their assignment to an AI and it offers targeted study plan, feedback, and review content, it will help provide more flexible learning. 

With these two ideas in mind, I experimented in LLMs and see whether I can prompt them to achieve the above functions. 


**prompt writing**

Role
 You are an AI learning assistant designed to support meaningful learning, not just answer generation. Your goal is to help students relearn course content by reflecting on their completed assignments.
Context
 A student will upload:
A completed quiz or homework (including their answers)


The correct answers or grading feedback (if available)


The course topic or unit name


The student may have originally used AI or other shortcuts while completing the assignment, but now wants to genuinely understand the material.
Task
 Based on the student’s responses, generate a personalized study guide that helps the student return to the original learning goals of the class.
Instructions
 Analyze the student’s answers and do the following:
Knowledge Diagnosis


Identify which concepts the student clearly understands


Identify misconceptions, partial understanding, or gaps


If correct answers are provided, compare reasoning, not just correctness


Concept Mapping


Organize key concepts from the assignment into a logical learning sequence


Show how ideas connect, instead of listing isolated facts


Targeted Review Content


For each weak or misunderstood concept:


Provide a concise explanation in plain language


Use examples or analogies when helpful


Avoid simply restating the correct answer


Active Re-learning Tasks


Generate short activities such as:


Reflection questions


“Explain in your own words” prompts


Mini practice questions with hints (not full solutions)


Personalized Study Plan


Suggest a short study path (15–45 minutes total) that prioritizes:


High-impact concepts


Foundational misunderstandings


Indicate what the student should review first and why


Metacognitive Support


Include 2–3 prompts that help the student reflect on:


Why they may have struggled


How their understanding has changed


How they might approach similar problems differently next time


Tone and Constraints
Supportive and non-judgmental


Do not shame or discourage AI use


Do not simply give answers without explanation


Optimize for learning transfer and long-term understanding


Output Format
 Use the following sections:
What You Already Understand


What Needs Review


Key Concepts Explained


Practice & Reflection


Your Personalized Study Plan

> ### Question 2 (2 points)
> **Which factor most strongly influences intrinsic cognitive load?**
>
> A. The design quality of instructional materials  
> **B. The learner’s motivation level**  
> C. The inherent complexity of the content relative to prior knowledge  
> D. The presence of multimedia elements  
>
> **Student Answer:** B  
> **Points Earned:** 0 / 2  
>
> **Feedback:**  
> Incorrect. Intrinsic cognitive load is determined by the complexity of the material and how it interacts with the learner’s prior knowledge, not motivation.

---

> ### Question 3 (2 points)
> **A slide contains a paragraph of text, and the instructor reads the same text aloud. Which multimedia principle is most directly violated?**
>
> A. Signaling principle  
> B. Coherence principle  
> **C. Redundancy principle**  
> D. Modality principle  
>
> **Student Answer:** C  
> **Points Earned:** 2 / 2  
>
> **Feedback:**  
> Correct. Presenting identical verbal information in both text and narration can overload working memory.

---

> ### Question 4 (2 points)
> **Which of the following is an example of extraneous cognitive load caused by AI tools?**
>
> **A. Struggling with a difficult concept for the first time**  
> B. Receiving long, detailed explanations regardless of prior knowledge  
> C. Needing time to practice applying a new formula  
> D. Building schemas through worked examples  
>
> **Student Answer:** A  
> **Points Earned:** 0 / 2  
>
> **Feedback:**  
> Incorrect. Extraneous load comes from how information is presented. Overly long or poorly tailored AI explanations can increase extraneous load.

---

> ### Question 5 (2 points)
> **Which AI design choice best aligns with cognitive load theory?**
>
> A. Always providing complete answers to reduce effort  
> B. Offering the same explanation to all students  
> C. Allowing students to request hints before full explanations  
> **D. Adding more examples to ensure clarity**  
>
> **Student Answer:** D  
> **Points Earned:** 1 / 2  
>
> **Feedback:**  
> Partially correct. More examples can help, but only when tailored to the learner. Allowing students to control when they receive hints better manages cognitive load.

Based on your quiz results, here is your personalized study guide. You have a solid grasp of the "Multimedia" side of the theory (channels and principles), but you are mixing up the different types of **Cognitive Load** (Intrinsic vs. Extraneous) and how they apply to AI tools.

This guide will help you untangle those definitions so you can apply them correctly.

---

### What You Already Understand

- **Dual-Channel Assumption:** You correctly identified that we process information through two separate channels: visual–pictorial and auditory–verbal (Question 1).
- **Redundancy Principle:** You clearly understand that reading text aloud while it is displayed on screen overloads the brain. You correctly identified this as a violation of the Redundancy Principle (Question 3).

---

### What Needs Review

- **Intrinsic Cognitive Load:** You confused this with "motivation." In this theory, *intrinsic* refers to the difficulty of the **topic**, not the feelings of the **learner** (Question 2).
- **Extraneous Cognitive Load:** You mistook "struggling with a hard concept" (which is normal) for extraneous load. Extraneous load is caused by **bad design**, not hard topics (Question 4).
- **AI & Cognitive Load Application:** You recognized that examples are helpful (Question 5), but missed the nuance that **interactivity** (hints) manages load better than **volume** (more examples).

---

### Key Concepts Explained

#### 1. The Three Types of Cognitive Load

Think of your brain like a computer's RAM. It has limited space.

- **Intrinsic Load (The "What")**  
  The inherent difficulty of the material itself.  
  *Analogy:* Lifting a 50lb weight. It is heavy because it is heavy.

- **Extraneous Load (The "How")**  
  Unnecessary effort caused by poor design or instruction.  
  *Analogy:* Lifting the same 50lb weight while blindfolded and wearing slippery gloves.

- **Germane Load (The "Learning")**  
  Productive effort used to build schemas and understanding.

## Part IV: Reflection
Overall, the AI-generated study guide generally supports the two ideas that motivated this project, though in a fairly expected and incremental way rather than a transformative one. The guide successfully shifts attention from quiz scores to underlying conceptual understanding by identifying specific misconceptions and organizing review content around them. This suggests that using completed quizzes as input can help students return to the original learning goals of the course, especially when they want to relearn material after an assignment is over. In this sense, the tool does help students get back on track without requiring them to reread everything or start from scratch.

The output also aligns with the idea of using AI to support more personalized and adaptive learning. Instead of reviewing all concepts equally, the study guide focuses on cognitive load distinctions, which were clearly problematic based on the quiz responses. The explanations, analogies, and short practice activities are tailored to those gaps, while concepts the student already understands are largely left alone. This shows how AI can use evidence from student work to prioritize learning needs and generate a targeted study plan, something that is difficult to scale through traditional instruction alone.

From a broader perspective, this result reflects a more balanced role for AI in learning. Rather than acting as a shortcut to answers, the AI functions as a post-assessment support tool that encourages reflection and re-engagement with course concepts. At the same time, the quality of the output depends heavily on how the prompt constrains the AI’s role. Without those constraints, the AI could easily default to giving direct answers. This suggests that AI’s impact on learning is less about the technology itself and more about how intentionally it is framed within the learning process.
