# A thought on AI Tutor: How may LLMs facilitate collaboration? 
In a previous post, I shared about gpt 5.2 features including images and video directly in chat responses. In this blog, I will explore the collaboration function in Chatgpt.

When I started the collaborated conversation, I did not assign chatgpt any role or give it any instruction, I just invited my other account in the conversation, and then I pretended that we were talking about next steps in our group presentation. 

I soon found out that Chatgpt will respond everytime anyone says anything, so it does not distinguish between whether users were talking to each other or if they need Chatgpt input. To avoid it keep doing this, I had to give it instruction of “@ChatGPT Can you only talk when we mention you?”. This was pretty effective. Chatgpt stopped responding soon. Then, when I said “@ChatGPT do you know?”, it started to give out answers. 

It also gave recommendations on how to split work: “If you want a concrete task split: one of you rewrites headlines + insight statements, the other scans for where reviewers might ask “but is this realistic?” and patches those spots.”

![Collaboration in ChatGPT](collab.png)

Thinking of this function in application, I think it will help greatly on collaborative learning. One of the criticisms of AI for learning is that it only allows 1 on 1 interaction with AI. However, now with the collaboration function, AI can facilitate group dynamics. First of all, by having an AI facilitator, group projects can be done easier. One push back or frustration for students to complete group projects is that groups do not work efficiently when some people do not respond or when no one is willing to share inspiring ideas. AI chatbot can jump in group chats, prompt people and perhaps even provide fairer grading for group projects. For example, teacher may send a prompt to AI to have it evaluate students’ engagement. I am looking forward for prompt that makes use of collaboration in Chatgpt!
